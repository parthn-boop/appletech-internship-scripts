```python
# Total sales by category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
# Total profit by category
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
# Set figure size for plot
plt.figure(figsize=(8,4))
```

Which categories generate highest sales?
Which categories generate highest profit?
Are high sales always high profit?
Which subcategories underperform?


```python
# Bar chart of total sales by category
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title("Category Sales")
```

Category sales chart

- These are 3 product categories each one shows how much money they made from sales.
- Furniture and Office supply has slightly lower sales as compare to technology.



```python
# Bar chart of total profit by category
plt.figure(figsize=(8,4))
sns.barplot(x=category_profit.index, y=category_profit.values)
plt.title("Category Profit")
```

Profit comparison chart


- Technology products make the most profit.
- Furniture products make the least profit.
- In furniture the sales are high but the profit is very low.




```python
# Boxplot of profit distribution by category
sns.boxplot(x='Category', y='Profit', data=df)
plt.title("Profit Distribution by Category")
```

Profit distribution boxplot

- Each dot shows profit of one order.
-  There are 3 product categories each one shows how profits are spread.
- The dot which are very high and down are profits that are very high or very low.
- If the category has very high profit then the category has also very high loss.


```python
# Top 10 customers by sales
top_sales = df.groupby('Customer Name')['Sales'].sum().nlargest(10)
# Top 10 customers by profit
top_profit = df.groupby('Customer Name')['Profit'].sum().nlargest(10)
```

```python
# Display top 10 customers by sales and profit
print("Top 10 Customers by Sales:\n", top_sales)
print("\nTop 10 Customers by Profit:\n", top_profit)
```

Top 10 customers by revenue?
Top 10 customers by profit?

```python
# Bar chart of top 10 customers by sales
plt.figure(figsize=(10,4))
sns.barplot(x=top_sales.index, y=top_sales.values, palette="Blues_d")
plt.title("Top 10 Customers by Sales")
plt.xticks(rotation=45)
```

- These bar shows how much money customers spent.
- Sean Miller has the highest sales.
- Sanjit Engle and Christopher Conant have lower sales.
- Basically the taller the bar the more sales the customer made.



```python
# Bar chart of top 10 customers by profit
plt.figure(figsize=(10,4))
sns.barplot(x=top_profit.index, y=top_profit.values, palette="Greens_d")
plt.title("Top 10 Customers by Profit")
plt.xticks(rotation=45)
plt.show()
```

- These bar shows how much profit customers made.

```python
# Total profit by state and city
state_profit = df.groupby('State')['Profit'].sum().sort_values(ascending=False)
city_profit = df.groupby('City')['Profit'].sum().sort_values(ascending=False)
# Bar chart of top 10 states by profit
plt.figure(figsize=(10,5))
sns.barplot(x=state_profit.index[:10], y=state_profit.values[:10])
plt.title("Top 10 States by Profit")
```

- These bar shows how much profit business made in each state.
- California and New York make the most profit.
- Delaware and Minnesota earn the least profit.


```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
# Group by month and calculate total sales & profit
monthly = df.groupby(df['Order Date'].dt.to_period('M')).agg({'Sales':'sum','Profit':'sum'})
# Plot monthly sales and profit trends
monthly.plot(subplots=True, figsize=(10,5), title="Monthly Sales & Profit Trends")
```

Monthly sales trend.
Monthly profit trend.
Seasonal patterns.
Peak business periods.

- The blue line shows how your sales changed month by month.And when line goes up the sales will increase.
- The orange line shows how your profit changed month by month.And when line goes up the profit will increase.
- The x axis shows the time when it is starting as it is starting from 2014 to 2017.
- The y-axis shows the money .
